### Bertscore for pegasus-x-large-booksum-1

RobertaTokenizer and RobertaModel are components of the RoBERTa (Robustly Optimized BERT Pretraining Approach) model, which is a variant of the popular BERT (Bidirectional Encoder Representations from Transformers) model. RoBERTa is built on the Transformer architecture and is specifically designed for pretraining language representations.

RobertaTokenizer is responsible for tokenizing raw text into subword units called tokens. It splits the input text into a sequence of tokens that can be understood by the model. The tokenizer also handles special tokens such as [CLS] (used for classification tasks) and [SEP] (used to separate sentences or segments).

RobertaModel is the core component that performs the actual language modeling. It takes the tokenized input sequence and generates contextualized representations for each token. The model consists of multiple layers of self-attention mechanisms and feed-forward neural networks. During training, the model learns to predict missing or masked tokens based on the surrounding context, which enables it to capture rich semantic representations.

RobertaTokenizer and RobertaModel are essential components of the RoBERTa model that enable tokenization and contextualized language representation learning, respectively, making them valuable tools for various natural language processing tasks

BERTScore computes a sentence-level similarity score by making use of token-level similarities, produced by cosine similarities between contextual embeddings. The numerical range of BERTScore is between -1 and 1, the same as the underlying cosine similarity. In practice, however, BERTscore is usually found to be in a small range. For an extreme case, BERTScore computed with the large RoBERTa model often is between 0.85 and 0.95.

```python
import pandas as pd
from transformers import RobertaTokenizer, RobertaModel
import torch

# Assuming you have a CSV file named 'your_data.csv' with 1620 rows and columns: 'Abstract', 'Claims', and 'Summary'

# Read only the first 1620 rows from the CSV file
df = pd.read_excel('pegasus-x-large-booksum-1.xlsx', nrows=1630)

# Extract the desired columns
columns_to_keep = ['Filename', 'Abstract', 'Claims', 'Summary']  # Add other column names here
df_subset = df[columns_to_keep]

# Store the 'Abstract' and 'Claims' columns in one variable
abstract_and_claims = df['Abstract'] + ' ' + df['Claims']

# Store the 'Summary' column in another variable
summary = df['Summary']

# Load tokenizer and model
tokenizer = RobertaTokenizer.from_pretrained('roberta-large')
model = RobertaModel.from_pretrained('roberta-large', output_hidden_states=True)

# Define maximum sequence length
max_length = 512

# Create a new DataFrame to store the BERT scores
bert_scores_df = pd.DataFrame()

for i in range(len(abstract_and_claims)):
    input_content = abstract_and_claims[i]
    reference_summary = summary[i]

    # Split input content and reference summary into segments (e.g., paragraphs)
    input_segments = input_content.split("\n")
    reference_segments = reference_summary.split("\n")

    # Initialize the scores for this row
    row_scores = {}

    for input_seg, reference_seg in zip(input_segments, reference_segments):
        # Tokenize input
        input_tokens = tokenizer.encode(input_seg, add_special_tokens=True, truncation=True, max_length=max_length)
        reference_tokens = tokenizer.encode(reference_seg, add_special_tokens=True, truncation=True, max_length=max_length)

        # Convert tokens to tensors
        input_ids = torch.tensor(input_tokens).unsqueeze(0)
        reference_ids = torch.tensor(reference_tokens).unsqueeze(0)

        # Generate attention masks
        input_attention_mask = torch.ones(input_ids.shape)
        reference_attention_mask = torch.ones(reference_ids.shape)

        # Calculate BERT scores
        with torch.no_grad():
            input_outputs = model(input_ids, input_attention_mask)
            reference_outputs = model(reference_ids, reference_attention_mask)

        # Get the last layer hidden states
        input_hidden_states = input_outputs.hidden_states[-1]
        reference_hidden_states = reference_outputs.hidden_states[-1]

        # Replace this with your desired scoring method or calculation
        scores = (input_hidden_states.mean() + reference_hidden_states.mean()) / 2

        row_scores.setdefault('scores', []).append(scores.item())

    # Calculate the average scores for this row
    num_segments = len(input_segments)
    row_scores_avg = {
        'scores': sum(row_scores['scores']) / num_segments
    }

    # Convert the row_scores_avg to a DataFrame and append to bert_scores_df
    row_scores_df = pd.DataFrame(row_scores_avg, index=[i])
    bert_scores_df = pd.concat([bert_scores_df, row_scores_df], ignore_index=True)

# Merge the original data with the BERT scores
merged_df = pd.concat([df_subset, bert_scores_df], axis=1)

# Save the scores to a CSV file
merged_df.to_csv('bert_scores_Final.csv', index=False)
```
