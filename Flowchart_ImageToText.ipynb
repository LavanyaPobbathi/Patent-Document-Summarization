```python
import cv2
from google.colab.patches import cv2_imshow
import easyocr

# Load the image
image_path = 'image_P1.jpeg'
image = cv2.imread(image_path)

# Initialize the OCR reader
reader = easyocr.Reader(lang_list=['en'])

# Perform text extraction
results = reader.readtext(image)

# Iterate through the detected text and print
for (bbox, text, prob) in results:
    print(f"Text: {text}, Probability: {prob:.2f}")

# Display the image
cv2_imshow(image)
```

```python
import os
from PIL import Image
import pytesseract

def main(image_path):
    if os.path.exists(image_path):
        filename = os.path.splitext(os.path.basename(image_path))[0]

        extracted_text = pytesseract.image_to_string(Image.open(image_path))
        print("Extracted Text:")
        print(extracted_text)

    else:
        print("Image not found at the given location")

if __name__ == '__main__':
    image_path = '/content/image_P1.jpg'  # Update with the path to your image
    image_path = os.path.abspath(image_path)
    main(image_path)
```
